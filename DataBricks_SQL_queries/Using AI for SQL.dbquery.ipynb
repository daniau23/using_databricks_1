```sql
SELECT 
  county_state_name, 
  population, 
  -- Remove commas and cast emissions to DOUBLE
  CAST(REPLACE(`GHG emissions mtons CO2e`, ',', '') AS DOUBLE) 
    -- Remove commas, cast population to DOUBLE, use NULLIF to avoid division by zero
    / NULLIF(CAST(REPLACE(population, ',', '') AS DOUBLE), 0) AS emissions_per_person
FROM emissions_data_2023
 ORDER BY emissions_per_person DESC
 LIMIT 10
```
